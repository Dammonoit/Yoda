```python
import requests
#import request python module
import bs4                 
from bs4 import BeautifulSoup  #importing beautiful soup

ActualUrl="https://www.stackoverflow.com"
#this is official stackoverflow link

# This is the static section of the url.
url="https://stackoverflow.com/search?q="
question=input("Enter your question : ") 
#question as query.
url=url+question.replace(" ","+")        
#addiing question to url and resplacing whitesapces with +
print('url: ',url)
response = requests.get(url)
#add response to get url
page=response.text
#store tha page in the test formatting.
print(page)
```

```python
#now make a soup of the page using beautiful soup library to access it's method.
soup = BeautifulSoup(page,"lxml")
initial_class=soup.findAll('div',{'class':'question-summary search-result'})
#initial class stores all values from question search.
vote_class=soup.findAll('span',{'class':'vote-count-post'})
#votes class stores all the votes in the page.
```

```python
print('initial_class: ',initial_class)
print('#'*50)
print('vote_class: ',vote_class)
print('#'*50)
print('type(initial_class): ',type(initial_class))
print('#'*50)
print('type(vote_class): ',type(vote_class))
```

```python
#define a list where no of votes will be stored.
NoOfVotes=list()
for i in range(len(vote_class)):
        NoOfVotes.append(int(vote_class[i].string))
print(NoOfVotes)
```

```python
#now we need to make lists of questions in the question lists.
questions=list()
for i in soup.findAll('a'):
    if i.get('data-searchsession') and i.get('href'):
        questions.append(i.get('title'))
```

```python
#let's store lists of numbers in qusetions.
question_number=[i for i in range(1,(len(initial_class)+1))]
```

```python
#now store all the links in a new list.
Link=list()
for link in soup.findAll('a'):
	    if link.get('data-searchsession') and link.get('href'):
		    Link.append("https://www.stackoverflow.com"+link.get('href'))
```
